```python
import tkinter as tk

def submit_data():
    username = username_entry.get()
    password = password_entry.get()
    course = course_var.get()
    state = state_var.get()
    languages = [lang.get() for lang in language_vars if lang.get()]  # Get selected languages

    # Do something with the collected data (e.g., print it)
    print(f"Username: {username}")
    print(f"Password: {password}")
    print(f"Course: {course}")
    print(f"State: {state}")
    print(f"Languages: {languages}")

root = tk.Tk()
root.title("Form")

# Labels
username_label = tk.Label(root, text="Username:")
password_label = tk.Label(root, text="Password:")
course_label = tk.Label(root, text="Course:")
state_label = tk.Label(root, text="State:")
language_label = tk.Label(root, text="Language:")

# Entries
username_entry = tk.Entry(root)
password_entry = tk.Entry(root, show="*")  # Hide password

# Course radio buttons
course_var = tk.StringVar(value=None)
btech_radio = tk.Radiobutton(root, text="BTech", variable=course_var, value="BTech")
mtech_radio = tk.Radiobutton(root, text="MTech", variable=course_var, value="MTech")

# State dropdown
state_var = tk.StringVar(value="MP")
state_dropdown = tk.OptionMenu(root, state_var, "MP", "Chhattisgarh", "UP", "Gujarat", "Kerala")

# Language checkboxes
language_vars = [tk.StringVar() for _ in range(3)]
language_checkboxes = [
    tk.Checkbutton(root, text="English", variable=language_vars[0], onvalue="english", offvalue="")
]
language_checkboxes.append(tk.Checkbutton(root, text="Hindi", variable=language_vars[1], onvalue="hindi", offvalue=""))
language_checkboxes.append(tk.Checkbutton(root, text="Other", variable=language_vars[2], onvalue="other", offvalue=""))

# Buttons
submit_button = tk.Button(root, text="Submit", command=submit_data)
reset_button = tk.Button(root, text="Reset")

# Grid layout
username_label.grid(row=0, column=0)
username_entry.grid(row=0, column=1)
password_label.grid(row=1, column=0)
password_entry.grid(row=1, column=1)
course_label.grid(row=2, column=0)
btech_radio.grid(row=2, column=1)
mtech_radio.grid(row=2, column=2)
state_label.grid(row=3, column=0)
state_dropdown.grid(row=3, column=1)
language_label.grid(row=4, column=0)
for i, checkbox in enumerate(language_checkboxes):
    checkbox.grid(row=4, column=i+1)
submit_button.grid(row=5, column=0)
reset_button.grid(row=5, column=1)

root.mainloop()
```
